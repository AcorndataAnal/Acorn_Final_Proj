```python
file= open('files/remain_unique_menus.txt','r')
data= file.read()
data_list= data.split('\n')
```

```python
def check_string_contains(string, *targets):
    for target in targets:
        return target in string
```

```python
result=[]
def check_words(word):
    for data in data_list:
        if check_string_contains(data, word) ==True:
            result.append(data)
    return result

def second_check(word):
    second=[]
    for data in result:
        if check_string_contains(data, word)== True:
            second.append(data)
    return second

check_words('열무')
열무나물= second_check('나물')
열무무침= second_check('무침')
```
