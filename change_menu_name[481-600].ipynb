```python
# 급식메뉴 중 식품영양소DB에 없는 음식
with open('files/remain_unique_menus.txt','r') as file:
    data= file.read()
    data_list= data.split('\n')
```

```python
# string 안에 target단어가 포함되어있으면 True 반환
def check_word(string, target):
    return target in string
```

```python
# 1차 탐색. data_list 중 word가 포함되어있는 음식
def basic(word):
    global basic_list
    basic_list=[]
    for menu in data_list:
        if check_word(menu, word) ==True:
            basic_list.append(menu)
    return basic_list

# 심층 탐색. 탐색한 결과 중 word가 포함되어있는 음식
def search(basic, word):
    global search_list
    search_list=[]
    for menu in basic:
        if check_word(menu, word)== True:
            search_list.append(menu)
    return search_list
```
