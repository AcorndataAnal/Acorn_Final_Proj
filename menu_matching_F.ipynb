```python
def check_contain(x,y,z=[]) :
    menu = []
    if len(z) == 0 :
        menu = [word for word in y 
                if all(keyword in word for keyword in x)]
    else :            
        menu = [word for word in y 
                if all(keyword in word for keyword in x) and
               all(keyword not in word for keyword in z)]
    return menu
```

```python
#### target 확인 = 내 범위에 있는 음식
x_t = ['샌드위치']; y_t = target
res_t = check_contain(x_t, y_t) ; print(res_t)
```
