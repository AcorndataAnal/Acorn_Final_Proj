```python
dict={} #  {include: [remain,..], include:[remain,..], ... } 
```

```python
# 식품영양DB 중 급식메뉴에 존재하는 음식
cat=[]
with open("../files/include_db_menus.txt", "r", encoding='utf-8') as f:
    for line in f:
        cat.append(line.split(":")[0])
```

```python
# 급식메뉴 중 식품영양소DB에 없는 음식
with open('../files/remain_unique_menus.txt','r',encoding='utf-8') as file:
    data= file.read()
    data_list= data.split('\n')
```

```python
# target단어가 string 안에 포함되어있으면 True 반환
def check_word(string, target):
    return target in string
```

```python
# 1차 탐색. data_list 중 word가 포함되어있는 음식
def basic(word):
    global basic_list
    basic_list=[]
    for menu in data_list:
        if check_word(menu, word) ==True:
            basic_list.append(menu)
    return basic_list

# 심층 탐색. 탐색한 결과 중 word가 포함되어있는 음식
def search(basic, word):
    global search_list
    search_list=[]
    for menu in basic:
        if check_word(menu, word)== True:
            search_list.append(menu)
    return search_list
```

```python
# ex1) 열무나물무침
열무= basic('열무') # 전제 급식메뉴 중 열무가 들어간 음식
len(열무)
```

```python
열무나물= search(열무,'나물') # 열무가 들어간 급식 중 나물이 들어간 음식
print(len(열무나물))
print(열무나물)
```

```python
열무무침= search(열무, '무침') # 열무가 들어간 급식 중 무침이 들어간 음식
print(len(열무무침))
열무무침
```

```python
열무나물무침= search(열무나물,'무침') # 열무와 나물이 들어간 급식 중 무침이 들어간 음식
print(len(열무나물무침))
열무나물무침
```

```python
cat121= list(set(열무나물+열무무침)) # set으로 중복 제거
len(cat121)
```

```python
dict[cat[121]]=cat121 # dict에 추가
dict
```
